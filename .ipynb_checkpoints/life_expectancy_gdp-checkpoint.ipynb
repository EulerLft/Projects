```python
import pandas as pd 
from matplotlib import pyplot as plt 
import seaborn as sns 

all_data = pd.read_csv('all_data.csv')
all_data
```

```python
chile_data = all_data[all_data['Country'] == 'Chile']
print('Chile data: ', len(chile_data))
print('Chile years: \n', chile_data['Year'].unique())


china_data = all_data[all_data['Country'] == 'China']
print('\nChina data: ', len(china_data))
print('China years: \n', china_data['Year'].unique())

germany_data = all_data[all_data['Country'] == 'Germany']
print('\nGermany data: ', len(germany_data))
print('Germany years: \n', germany_data['Year'].unique())

mexico_data = all_data[all_data['Country'] == 'Mexico']
print('\nMexico data: ', len(mexico_data))
print('Mexico years: \n', mexico_data['Year'].unique())

usa_data = all_data[all_data['Country'] == 'United States of America']
print('\nUSA data: ', len(usa_data))
print('USA years: \n', usa_data['Year'].unique())

zimbabwe_data = all_data[all_data['Country'] == 'Zimbabwe']
print('\nZimbabwe data: ', len(zimbabwe_data))
print('Zimbabwe years: \n', zimbabwe_data['Year'].unique())
```

```python
average_gdp = all_data.groupby('Country')['GDP'].mean().reset_index()

average_age = all_data.groupby('Country')['Life expectancy at birth (years)'].mean().reset_index()

countries = list(average_age['Country'])
countries_list = [item.replace('United States of America', 'USA') for item in countries]
```

```python
fig, ax = plt.subplots(1, 2, figsize=(14, 6))

ax[0].bar(average_gdp['Country'], average_gdp['GDP'])
ax[0].tick_params(axis='x', rotation=30);
ax[0].set_xticks(average_gdp['Country']);
ax[0].set_xticklabels(countries_list);
ax[0].set_ylabel('GDP in Trillions of USD')

ax[0].set_title('Average GDP by Country');

ax[1].bar(average_age['Country'], average_age['Life expectancy at birth (years)']);
ax[1].tick_params(axis='x', rotation=30);
ax[1].set_xticks(average_gdp['Country']);
ax[1].set_xticklabels(countries_list);
ax[1].set_ylabel('Age (yrs)');
ax[1].set_ylim(30, 85);
ax[1].set_title('Average Age by Country');
```

```python
fig, ax = plt.subplots(figsize=(10, 8))

plt.subplot(2,3,1)
plt.scatter(chile_data['Year'], chile_data['Life expectancy at birth (years)'], label='Chile');

plt.subplot(2,3,2)
plt.scatter(china_data['Year'], china_data['Life expectancy at birth (years)'], label='China');

plt.subplot(2,3,3)
plt.scatter(germany_data['Year'], germany_data['Life expectancy at birth (years)'], label='Germany');

plt.subplot(2,3,4)
plt.scatter(mexico_data['Year'], mexico_data['Life expectancy at birth (years)'], label='Mexico');

plt.subplot(2,3,5)
plt.scatter(usa_data['Year'], usa_data['Life expectancy at birth (years)'], label='USA');

plt.subplot(2,3,6)
plt.scatter(zimbabwe_data['Year'], zimbabwe_data['Life expectancy at birth (years)'], label='Zimbabwe');

plt.ylabel('Age');
plt.xlabel('Years');

plt.legend();
```
