```python
import numpy as np
import pandas as pd 
from matplotlib import pyplot as plt 
import seaborn as sns 

all_data = pd.read_csv('all_data.csv')
all_data.head()
```

```python
chile_data = all_data[all_data['Country'] == 'Chile']
china_data = all_data[all_data['Country'] == 'China']
germany_data = all_data[all_data['Country'] == 'Germany']
mexico_data = all_data[all_data['Country'] == 'Mexico']
usa_data = all_data[all_data['Country'] == 'United States of America']
zimbabwe_data = all_data[all_data['Country'] == 'Zimbabwe']
```

```python
list_years = list(all_data['Year'].unique())
years = []

for year in list_years:
    if year % 2 == 0:
        years.append(str(year))
    else:
        years.append('')
```

```python
fig, ax = plt.subplots(2, 3, figsize=(14, 8))

ax[0,0].scatter(chile_data['Year'], chile_data['Life expectancy at birth (years)'], label='Chile', color='gray');
ax[0,0].set_title('(Chile)')

ax[0,1].scatter(china_data['Year'], china_data['Life expectancy at birth (years)'], label='China', color='r');
ax[0,1].set_title('(China)')

ax[0,2].scatter(germany_data['Year'], germany_data['Life expectancy at birth (years)'], label='Germany', color='black');
ax[0,2].set_title('(Germany)')

ax[1,0].scatter(mexico_data['Year'], mexico_data['Life expectancy at birth (years)'], label='Mexico', color='tab:green');
ax[1,0].set_title('(Mexico)')

ax[1,1].scatter(usa_data['Year'], usa_data['Life expectancy at birth (years)'], label='USA', color='tab:blue');
ax[1,1].set_title('(USA)')

ax[1,2].scatter(zimbabwe_data['Year'], zimbabwe_data['Life expectancy at birth (years)'], label='Zimbabwe', color='tab:orange');
ax[1,2].set_title('(Zimbabwe)')

for a in ax.flatten():
    a.set_xticks(chile_data['Year'].unique())
    a.set_xticklabels(years)
    a.tick_params(axis='x', rotation=30)

plt.ylabel('Age');
plt.xlabel('Years');
plt.subplots_adjust(wspace=0.25, hspace=0.4)

fig.legend(loc='center right', bbox_to_anchor=(1.025, 0.5));
```
